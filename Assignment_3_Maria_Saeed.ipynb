```python
def check(board, player):
  #you need to check if the king is under attack
  #sabsay pehlay you check king kee moves
  opp_all_moves=set()
  opp=set()
  oppon=""
  if player=="W":
    oppon="B"
  else:
    oppon="W"


  pieces=["P1","P2","P3","P4","P5","P6","P7","P8","K","Q","R1","R2","B1","B2"]
  all_king_moves=set()
  #print("checking for checkmate1")
  all_king_moves,x,y=actions(board,"K", player)
  #print("checking for checkmate2")
  all_king_moves.add((x,y))
  #print("checking for checkmate3")
  for i in pieces:
    #abb we are in the pieces
     #print("checking for checkmate4")
     opp,x,y=actions(board, i ,oppon)
     if opp!=None:



      for k in opp:
        #print("checking for checkmate5")
        opp_all_moves.add(k)
  #now i check opponent keh saray moves agar opponent keh total moves main king keh saray moves hain along with his current position toh we
  #check mate
  s=all_king_moves.issubset(opp_all_moves)
  #print("checking for checkmate")
  #print(s)
  return  s
```

```python
def ab_pruning(board, depth, alpha, beta, MP):
    if depth == 0 or check(board,MP):
        ss=evaluate(board, MP)
        #print("evaluation score")
        #print(ss)
        return ss
    pieces=["P1","P2","P3","P4","P5","P6","P7","P8","K","Q","R1","R2","B1","B2"]
    if MP=="B":
        max_eval = float('-inf')

        for i in pieces:
          #actions(board,i, "B")
          act,x,y=actions(board,i, "B")
          for move in act:
              new_board = making_a_move(board, move,piece)
              eval = ab_pruning(new_board, depth - 1, alpha, beta, "W")
              max_eval = max(max_eval, eval)
              alpha = max(alpha, eval)
              if beta <= alpha:
                  break
        return max_eval
    else:
        min_eval = float('inf')
        for i in pieces:
          act,x,y=actions(board,i,"W")
          for move in act:

              new_board = making_a_move(board, move,piece)
              eval = ab_pruning(new_board, depth - 1, alpha, beta, "B")
              min_eval = min(min_eval, eval)
              beta = min(beta, eval)
              if beta <= alpha:
                  break
        return min_eval
```

```python
def making_a_move(board, move,piece):
  if  piece in board:
    ind=board.index(piece)
    (i,j)=ind
    board[i][j]="--"
    (i,j)=move
    board[i][j]=piece
  return board
```

```python
def result(board, action, piece,playerss,curr_x,curr_y):
    """
    Returns the board that results from making move (i, j) on the board.
    """

    p=playerss+piece
    result_board = copy.deepcopy(board)
    (i, j) = action
    result_board[i][j] = p
    #print("the currents")
    #print(curr_x)
    #print(curr_y)
    result_board[int(curr_x)][int(curr_y)]="--"
    return result_board
```
